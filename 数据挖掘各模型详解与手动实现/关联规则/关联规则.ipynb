```python
import mlxtend
```

```python
import pandas as pd 
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# 自定义数据集
data = {
    'ID':[1,2,3,4,5,6],
    'Oninon':[1,0,0,1,1,1],
    'Patoto':[1,1,0,1,1,1],
    'Burger':[1,1,0,0,1,1],
    'Milk':[0,1,1,1,0,1],
    'Beer':[0,0,1,0,1,0]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df_goods = df.drop('ID',axis=1)
```

```python
# 取支持度大于50%的商品
frequant_itemsets = apriori(df[df_goods.columns], min_support=0.50, use_colnames=True)
```

```python
frequant_itemsets
```

```python
# 计算关联规则
rules = association_rules(frequant_itemsets,metric='lift',min_threshold=1)
rules
```

```python
# 筛选提升度>1.125并且置信度 > 0.8的

rules[(rules['lift']>1.125) & (rules['confidence']>0.8)]
```
